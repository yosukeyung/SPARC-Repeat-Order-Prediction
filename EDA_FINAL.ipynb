## Pemformatan Data
Penyesuaian format ini wajib kita lakukan agar algoritma *machine learning* nantinya tidak keliru dalam memproses data, sehingga perhitungan matematis maupun visualisasi yang kita buat bisa benar-benar akurat.

Terlihat ada beberapa tipe data yang kurang cocok terbaca oleh sistem. Hal ini perlu kita rapikan terlebih dahulu sebelum mulai membuat visualisasi (EDA) ataupun mengisi sisa data yang kosong. Beberapa penyesuaian yang kita lakukan antara lain:
1. Kolom Cash/Credit dan Kode Dealer diubah tipe datanya menjadi string. Alasannya meskipun isinya angka, keduanya sebenarnya adalah kode kategori yang tidak boleh diolah secara matematis oleh model.
2. Pada kolom dp aktual dan cicilan, kita perlu menghapus teks "Rp", spasi, serta tanda titik. Tujuannya agar kolom ini bisa diubah murni menjadi data angka.
3. Setelah menjadi numerik, dilakukan koreksi terhadap data yang tidak masuk akal. Misalnya, jika nilai uang muka (DP) melebihi harga kendaraan (OTR), maka nilai DP akan disamakan dengan harga OTR (dianggap lunas). Sisa *outlier* ekstrem dipangkas (*capping*) pada persentil 99%.
4. Beberapa kolom dihapus untuk mengurangi dimensi data karena informasinya tumpang tindih. Kolom Tgl Lahir dan tgl mohon dihapus karena informasinya sudah terwakili oleh kolom umur. Kolom range dp juga dihapus karena rentangnya banyak yang tidak valid dan nilainya sudah diwakili oleh angka pasti pada dp aktual.

Langkah penyesuaian format sangat penting agar data kita siap dan valid saat dibentuk menjadi grafik pada tahap eksplorasi nanti.



```python
tipe_pembayaran = {
    1 : 'Cash',
    2 : 'Credit' #Membuat kamus untuk mapping nilai numerik pada kolom 'Cash/Credit' 
}

df['Cash/Credit'] = df['Cash/Credit'].map(tipe_pembayaran) 

df.head()
```

```python
#Mengubah tipe data jadi string karena isinya kode kategori agar tidak dihitung secara matematis, tapi tetap bisa dihitung frekuensinya saja
df['Kode Dealer'] = df['Kode Dealer'].astype(str)
```

```python
#Kolom yang berisi nilai uang, kita ubah ke numerik setelah dibersihkan dari simbol dan spasi
kolom_uang = ['dp aktual', 'cicilan'] 

for col in kolom_uang:
    df[col] = df[col].astype(str) #Memastikan semua data di kolom ini adalah string untuk manipulasi teks
    df[col] = df[col].str.replace('Rp', '', case=False) #Menghapus simbol 'Rp' jika ada, case-insensitive
    df[col] = df[col].str.replace(' ', '', regex=False) #Menghapus spasi yang mungkin ada di antara angka
    df[col] = df[col].str.replace('.', '', regex=False) #Menghapus titik yang mungkin digunakan sebagai pemisah ribuan
    df[col] = pd.to_numeric(df[col], errors='coerce') #Mengubah ke numerik, jika ada yang tidak bisa diubah akan menjadi NaN
```

```python
#Baris yang memiliki dp aktual dan cicilan sama dengan 0 kemungkinan adalah data yang tidak valid atau tidak lengkap, sehingga kita menghapusnya.
df = df[~((df['dp aktual'] == 0) & (df['cicilan'] == 0))].reset_index(drop=True)
```

```python
#Jika DP lebih besar dari OTR, maka dianggap sudah dibayar lunas (DP aktual = OTR)
df['dp aktual'] = np.where(df['dp aktual'] > df['OTR'], df['OTR'], df['dp aktual'])

#Jika DP aktual lebih besar dari 99 persentil, maka dianggap outlier, set ke batas atas persentil 99% untuk DP sekarang
batas_atas_dp = df['dp aktual'].quantile(0.99)
print(f"Batas atas persentil 99% untuk DP sekarang: Rp {batas_atas_dp:,.0f}")

df['dp aktual'] = np.where(df['dp aktual'] > batas_atas_dp, batas_atas_dp, df['dp aktual']) #mengganti nilai DP aktual yang lebih besar dari batas atas dengan batas atas itu sendiri

df
```

```python
#Menghapus tabel Tgl Lahir dan tgl mohon karena sudah ada tabel usia, informasi redundant
df = df.drop(columns=['Tgl Lahir', 'tgl mohon'])
```

```python
#Menghapus kolom kategori range dp karena sudah ada dp aktual yang lebih informatif, informasi redundant
df = df.drop(columns=['range dp'])
```

## Pengisian Data Kosong
Setelah data dipisahkan, kita menerapkan dua strategi pengisian yang berbeda agar informasi yang dihasilkan tetap akurat.

### Kategorikal
Untuk data berupa teks atau kategori, pengisian nilai kosong (NaN) disesuaikan dengan kondisi masing-masing kolom agar hasilnya lebih akurat:
1. Kolom Kelurahan dan kode motor yang kosong diisi dengan nilai yang paling sering muncul (modus), tetapi dikelompokkan berdasarkan Kecamatan dan type series-nya. Penggunaan nilai modus dipilih karena fitur tidak dapat dihitung secara matematis, sehingga mengambil nilai yang paling sering muncul adalah pendekatan statistik yang paling logis untuk mewakili kelompok tersebut.
2. Kolom Finance Company yang kosong diisi dengan huruf 'N'. Alasannya, data yang kosong ini berasal dari pelanggan yang membayar secara tunai (Cash), sehingga memang tidak ada perusahaan pembiayaan yang terlibat.
3. Kolom Pekerjaan yang kosoong diisi dengan huruf 'N'. Alasannya, data yang kosong ini dapat diisi dengan kategori 'N' yakni mempresentasikan Pekerjaan 'Lain-Lain'
3. Untuk sisa kolom kategori lainnya, nilai yang kosong langsung diisi secara serentak menggunakan modus dari masing-masing kolom, karena tidak ada faktor spesifik lain yang memengaruhinya.

```python
df['Kelurahan'] = df['Kelurahan'].fillna(df.groupby('Kecamatan')['Kelurahan'].transform(lambda x: x.mode().iloc[0])) 
df['kode motor'] = df['kode motor'].fillna(df.groupby('type series')['kode motor'].transform(lambda x: x.mode().iloc[0])) 
df['Finance Company']=df['Finance Company'].fillna('N') #Mengisi data kosong pada tabel finance company dengan kategori "Lain-lain"
df['Pekerjaan']=df['Pekerjaan'].fillna('N') #Mengisi data kosong pada tabel pekerjaan dengan kategori "Lain-lain"

#Mengecualikan kolom yang sudah bersih dari daftar agar tidak ikut diisi ulang
kol_kategorikal.remove('Gender')
kol_kategorikal.remove('Customer ID')
kol_kategorikal.remove('Kelurahan') 
kol_kategorikal.remove('kode motor')       
kol_kategorikal.remove('Finance Company')
kol_kategorikal.remove('Pekerjaan')

#Mengisi sisa kolom kategori yang masih kosong secara massal dengan nilai modus masing-masing kolom
df[kol_kategorikal] = df[kol_kategorikal].fillna(df[kol_kategorikal].mode().iloc[0]) 
```

### Numerikal
Nilai kosong (NaN) pada fitur angka diisi menggunakan nilai tengah atau median. Pendekatan ini dipilih karena median jauh lebih kebal terhadap nilai ekstrem (outlier) dibandingkan nilai rata-rata biasa (mean). Agar angka yang diisikan akurat dan relevan dengan kondisi aslinya, pengisian dilakukan secara spesifik berdasarkan kelompok data yang saling berkaitan:

1. Kolom OTR diisi menggunakan median dari motor dengan type series yang sama.
2. Kolom cicilan diisi menggunakan median berdasarkan kode motor yang sama. Hal ini menggambarkan kondisi di lapangan, di mana nominal cicilan pasti bergantung pada spesifikasi motor yang dipilih.
3. Sebagai langkah antisipasi, jika masih terdapat data yang kosong atau tidak memiliki kelompok spesifik, sisa data tersebut akan langsung diisi menggunakan nilai median keseluruhan dari masing-masing kolom.

```python

df['OTR'] = df['OTR'].fillna(df.groupby('type series')['OTR'].transform('median')) 
df['cicilan'] = df['cicilan'].fillna(df.groupby('kode motor')['cicilan'].transform('median')) 

#Mengisi sisa data yang masih kosong dengan median keseluruhan, berjaga-jaga jika ada kelompok data yang sebelumnya kosong total.
df['OTR'] = df['OTR'].fillna(df['OTR'].median()) 
df['cicilan'] = df['cicilan'].fillna(df['cicilan'].median())
df['dp aktual'] = df['dp aktual'].fillna(df['dp aktual'].median())
df['umur']=df['umur'].fillna(df['umur'].median())
```

# Pembuatan Label Prediksi (Target Engineering)
Tahap selanjutnya adalah membentuk variabel target yang nantinya akan diprediksi oleh model *machine learning*. Karena tujuan utama kita adalah memprediksi apakah seorang pelanggan akan melakukan pembelian kembali (*repeat order*), kita perlu mengambil informasi tersebut dari riwayat transaksi yang ada.

Langkah yang dilakukan pada tahap ini meliputi:
1. Kita menghitung total kemunculan setiap pelanggan (`Customer ID`) di dalam dataset. Jika seorang pelanggan memiliki lebih dari satu transaksi, maka mereka diberi label 'Yes' pada kolom target baru bernama `repeat_order`. Jika hanya satu kali, maka diberi label 'No'.
2. Agar model dapat belajar memprediksi peluang *repeat order* murni dari profil awal pelanggan (bukan profil dari transaksi ke-2 atau ke-3), kita harus membuang riwayat transaksi lanjutannya. Oleh karena itu, kita memfilter data dengan hanya mempertahankan baris transaksi pertama dari setiap pelanggan.
3. Kolom `total_transaksi` yang sebelumnya digunakan sebagai alat bantu hitung akan dihapus agar tabel tetap rapi dan tidak membebani model.

```python
df['total_transaksi'] = df.groupby('Customer ID')['Customer ID'].transform('count')
df['repeat_order'] = np.where(df['total_transaksi'] > 1, 'Yes', 'No')
```

```python
df = df.drop_duplicates(subset=['Customer ID'], keep='first').copy() #Menghapus duplikasi berdasarkan 'Customer ID', hanya menyimpan transaksi pertama untuk setiap pelanggan
df = df.drop(columns=['total_transaksi'])
```

# Deteksi Outlier dengan Box Plot

Setelah data dibersihkan, kita melakukan pengecekan sebaran data pada fitur-fitur numerik untuk mendeteksi adanya *outlier*. Visualisasi Box Plot sangat efektif digunakan karena dapat merangkum lima informasi statistik utama: nilai minimum, kuartil pertama (Q1), median (Q2), kuartil ketiga (Q3), dan nilai maksimum.

Melalui visualisasi ini, kita dapat:
1. Menemukan titik-titik data yang berada jauh di luar jangkauan (batas atas dan bawah), yang menandakan adanya anomali atau variasi data yang sangat lebar.
2. Menentukan fitur mana saja yang memerlukan penanganan khusus, seperti trimming atau clipping, agar tidak merusak performa model prediksi nantinya.
3. Melihat apakah data cenderung terkonsentrasi di satu area atau justru memiliki banyak variasi yang "berantakan".

```python
sns.set_theme(style="whitegrid")

#Mengatur tampilan titik outlier agar lebih mencolok
outlier_style = dict(markerfacecolor='red', marker='o', markersize=5, alpha=0.6)
```

```python
#Kode Visualisasi Boxplot untuk Deteksi Anomali pada Kolom Umur Pelanggan (before handling outlier)

plt.figure(figsize=(10, 3))
sns.boxplot(x=df['umur'], color='#3498db', flierprops=outlier_style, width=0.4)

plt.title('Deteksi Anomali: Distribusi Umur Pelanggan', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Umur (Tahun)', fontsize=12)
plt.tight_layout()
plt.show()
```

Temuan: Berdasarkan grafik distribusi umur pelanggan sebelum handling outlier, terlihat adanya rentetan outlier yang sangat ekstrem di sebelah kanan. Sumbu X memanjang secara tidak wajar hingga menyentuh angka lebih dari 1.000, yang mengindikasikan terekamnya usia pelanggan mencapai ratusan hingga ribuan tahun. Kondisi yang secara logis tidak mungkin ini bisa disebabkan oleh human error atau faktor lain sehingga perlu dibersihkan.

```python
#handling outlier pada kolom umur dengan mengganti nilai yang kurang dari 17 tahun atau lebih dari 100 tahun menjadi NaN, lalu mengisi NaN tersebut dengan median umur yang sudah dibersihkan dari outlier

df['umur'] = np.where((df['umur'] < 17) | (df['umur'] > 100), np.nan, df['umur'])
df['umur'] = df['umur'].fillna(df['umur'].median())

sns.boxplot(x=df['umur'], color='#3498db', flierprops=outlier_style, width=0.4)

plt.title('Deteksi Anomali: Distribusi Umur Pelanggan', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Umur (Tahun)', fontsize=12)
plt.tight_layout()
plt.show()
```

Pada grafik sesudah handling outlier, distribusi umur pelanggan sudah kembali normal. Perbaikan dilakukan dengan membatasi rentang usia logis antara 17 hingga 100 tahun. Batas bawah 17 tahun dipilih berdasarkan syarat minimal kepemilikan KTP, sedangkan batas atas 100 tahun ditetapkan sebagai filter keamanan untuk menghapus kesalahan input data yang ekstrem tanpa menghilangkan data pelanggan senior yang masih valid. Kita berasumsi bahwa angka 100 tahun merupakan batas usia maksimal yang masih memungkinkan secara biologis bagi seseorang untuk hidup dan terlibat dalam transaksi ekonomi. Meskipun batas atas ditetapkan pada 100 tahun, hasil visualisasi menunjukkan bahwa secara alami profil pelanggan kita berhenti di kisaran usia 60-an tahun, dengan mayoritas utama terpusat pada rentang usia produktif antara 20 hingga 40-an tahun.

```python
plt.figure(figsize=(10, 3))
sns.boxplot(x=df['OTR'], color='#f39c12', flierprops=outlier_style, width=0.4)

plt.title('Deteksi Anomali: Distribusi Harga OTR Kendaraan', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Harga On The Road / OTR (Rupiah)', fontsize=12)
plt.tight_layout()
plt.show()
```

Grafik boxplot sebelum penanganan menunjukkan adanya titik-titik anomali (outlier) yang sangat ekstrem di sebelah kanan. Nilai harga kendaraan (OTR) ini terlalu tinggi dan tidak wajar untuk standar harga sepeda motor. Kemungkinan besar, anomali ini terjadi akibat kesalahan input data, seperti kelebihan mengetik angka nol. Nilai ekstrem yang memanjang ini membuat kotak sebaran data utama (Q1-Q3) menjadi sangat pipih, sehingga mengaburkan rentang harga pasar yang sebenarnya.

```python
#Jika OTR lebih besar dari 99 persentil, maka dianggap outlier, set ke batas atas persentil 99% untuk OTR sekarang
batas_atas_otr = df['OTR'].quantile(0.99)
print(f"Batas atas persentil 99% untuk OTR: Rp {batas_atas_otr:,.0f}")

df['OTR'] = np.where(df['OTR'] > batas_atas_otr, batas_atas_otr, df['OTR'])

#Mengisi sisa data yang masih kosong dengan median keseluruhan, berjaga-jaga jika ada kelompok data yang sebelumnya kosong total.
median_otr = df['OTR'].median()
df['OTR'] = df['OTR'].fillna(median_otr)

print(f"Nilai NaN (harga tidak wajar) diisi median: Rp {median_otr:,.0f}")
```

```python
plt.figure(figsize=(10, 3))
sns.boxplot(x=df['OTR'], color='#f39c12', flierprops=outlier_style, width=0.4)

plt.title('Deteksi Anomali: Distribusi Harga OTR Kendaraan', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Harga On The Road / OTR (Rupiah)', fontsize=12)
plt.tight_layout()
plt.show()
```

Pada grafik sesudah penanganan outlier, sebaran harga OTR secara visual terlihat lebih banyak nilai ekstrimnya tetapi untuk logika bisnis ini masuk akal dan mencerminkan harga motor yang sebenarnya di pasaran. Perbaikan ini dilakukan dengan membatasi nilai harga yang ekstrem (capping) pada batas persentil 99% dan mengisi data yang kosong (NaN) dengan nilai median. Hasilnya, kotak sebaran data kini terlihat proporsional. Grafik ini membuktikan bahwa mayoritas transaksi pelanggan sudah terkonsentrasi di rentang harga yang valid, yaitu pada kisaran belasan hingga puluhan juta rupiah

```python
plt.figure(figsize=(10, 3))
sns.boxplot(x=df['dp aktual'], color='#e74c3c', flierprops=outlier_style, width=0.4)

plt.title('Deteksi Anomali: Distribusi DP Aktual', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Nominal DP (Rupiah)', fontsize=12)
plt.tight_layout()
plt.show()
```

Berdasarkan grafik boxplot sebelum handling outlier, sebaran data mayoritas pelanggan sebenarnya sudah berada di angka yang wajar. Namun, terdapat titik-titik anomali (outlier) di sebelah kanan yang nilainya sangat ekstrem hingga mencapai miliaran rupiah. Nilai DP yang jauh lebih besar dari harga motor ini sangat tidak masuk akal dan kemungkinan besar terjadi mungkin karena salah ketik saat input data (misalnya kelebihan angka nol). Data yang salah ini harus dibersihkan agar tidak mengganggu hasil prediksi algoritma nantinya.

```python
df['dp aktual'] = np.where(df['dp aktual'] > df['OTR'], df['OTR'], df['dp aktual'])

#Jika DP aktual lebih besar dari 99 persentil, maka dianggap outlier, set ke batas atas persentil 99% untuk DP sekarang
batas_atas_dp = df['dp aktual'].quantile(0.99)
print(f"Batas atas persentil 99% untuk DP sekarang: Rp {batas_atas_dp:,.0f}")

df['dp aktual'] = np.where(df['dp aktual'] > batas_atas_dp, batas_atas_dp, df['dp aktual'])

plt.figure(figsize=(10, 3))
sns.boxplot(x=df['dp aktual'], color='#e74c3c', flierprops=outlier_style, width=0.4)

plt.title('Deteksi Anomali: Distribusi DP Aktual', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Nominal DP (Rupiah)', fontsize=12)
plt.tight_layout()
plt.show()
```

Pada grafik boxplot setelah penanganan outlier, distribusi nilai DP Aktual tidak berubah sama sekali. Proses pembersihan dilakukan dalam dua tahap. Pertama, jika terdapat nilai DP yang melebihi harga unit (OTR), maka nilainya disamakan dengan harga OTR karena dianggap sebagai pembayaran lunas. Kedua, nilai-nilai yang masih terlalu ekstrem dibatasi menggunakan metode capping pada persentil 99%. Hasilnya, tidak ada data yang melebihi batas atas persentil sehingga boxplot tidak berubah sama sekali.

```python
plt.figure(figsize=(10, 3))
sns.boxplot(x=df['cicilan'], color='#2ecc71', flierprops=outlier_style, width=0.4)

plt.title('Deteksi Anomali: Distribusi Cicilan per Bulan', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Nominal Cicilan (Rupiah)', fontsize=12)
plt.tight_layout()
plt.show()
```

Grafik boxplot sebelum penanganan menunjukkan adanya rentetan titik anomali (outlier) ekstrem di sebelah kanan. Nilai cicilan per bulan ini terlihat tidak wajar karena ada yang mencapai angka besar untuk ukuran angsuran sepeda motor. Selain indikasi kesalahan input data, terdapat juga ketidakkonsistenan logika di mana pelanggan yang sudah membayar lunas (DP = OTR) namun masih tercatat memiliki tagihan cicilan.

```python
#Mengecek berapa banyak baris yang memiliki DP aktual sama dengan OTR (lunas) tapi cicilannya masih lebih besar dari 0 (seharusnya tidak ada cicilan jika sudah lunas)
lunas_tapi_nyicil = df[(df['dp aktual'] == df['OTR']) & (df['cicilan'] > 0)] 
print(f"Jumlah anomali (Lunas tapi ada cicilan): {len(lunas_tapi_nyicil)} baris")

#Jika dp aktual == OTR, maka cicilan paksa jadi 0. jika tidak, biarkan nilai cicilan apa adanya.
df['cicilan'] = np.where(df['dp aktual'] == df['OTR'], 0, df['cicilan'])
df['cicilan'] = np.where(df['cicilan'] > df['OTR'], np.nan, df['cicilan'])

df['cicilan']=df['cicilan'].fillna(df['cicilan'].median())

#Jika kolom 'Cash/Credit' ada di DataFrame, maka lakukan validasi silang untuk memperbaiki kategori pembayaran berdasarkan kondisi DP aktual dan OTR
if 'Cash/Credit' in df.columns: 
    df['Cash/Credit'] = np.where(df['dp aktual'] == df['OTR'], 'Cash', 'Credit')
    
df.loc[df['Cash/Credit'] == 'Cash', 'Tenor'] = 'N' #Memastikan bahwa jika pembayaran tunai, maka tenor otomatis jadi 'N' (Non-Installment)

plt.figure(figsize=(10, 3))
sns.boxplot(x=df['cicilan'], color='#2ecc71', flierprops=outlier_style, width=0.4)

plt.title('Deteksi Anomali: Distribusi Cicilan per Bulan', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Nominal Cicilan (Rupiah)', fontsize=12)
plt.tight_layout()
plt.show()
```

Pada grafik sesudah penanganan, sebaran data cicilan terlihat lebih banyak outliernya karena ketika outlier ekstrim dihapus maka batas whisker makin sempit yang membuat lebih banyak data dibaca outlier. Perbaikan dilakukan dengan memastikan konsistensi data. Setiap transaksi yang lunas dipaksa memiliki nilai cicilan nol dan status pembayaran "Cash". Selain itu, nilai cicilan yang tidak masuk akal (melebihi harga motor) telah dibersihkan dan diganti dengan nilai median. Hasilnya, kotak distribusi data kini terlihat proporsional tanpa ada lagi gangguan dari pencilan ekstrem. 

Kita rapikan ulang ketentuan dari Cash/Credit yang bernilai Credit harus memiliki informasi Finance Company dan Tenor, sehingga data yang bernilai 'N' akan kami isi dengan modus dari masing masing kolomnya

```python
kondisi_kredit_bocor = (df['Cash/Credit'] == 'Credit') & (df['Finance Company'] == 'N') & (df['Tenor'] == 'N')

# Isi dengan Modus (Nilai terbanyak)
modus_leasing = df[df['Finance Company'] != 'N']['Finance Company'].mode()[0]
modus_tenor = df[df['Tenor'] != 'N']['Tenor'].mode()[0]

df.loc[kondisi_kredit_bocor, 'Finance Company'] = modus_leasing
df.loc[kondisi_kredit_bocor, 'Tenor'] = modus_tenor
```

## Multivariat
Tahapan ini digunakan untuk mengamati interaksi antar banyak fitur sekaligus. Dengan menggunakan bantuan matriks korelasi (heatmap), kita dapat memahami hubungan data yang lebih luas antar seluruh variabel numerik di dalam dataset.

```python
plt.figure(figsize=(8, 6))
kolom_numerik = ['umur', 'dp aktual', 'cicilan', 'OTR']
matriks_korelasi = df[kolom_numerik].corr()

#Menambahkan masker untuk menyembunyikan angka yang dobel/redundant di segitiga atas
masker = np.triu(np.ones_like(matriks_korelasi, dtype=bool))

#Membuat heatmap untuk menampilkan matriks korelasi antar fitur numerik dengan nilai korelasi di dalam kotak
sns.heatmap(matriks_korelasi, mask=masker, annot=True, cmap='coolwarm', fmt=".2f", linewidths=1, linecolor='white', cbar_kws={"shrink": .8})  

plt.title('Heatmap Korelasi Fitur Numerik')
plt.show()
```

Visualisasi Heatmap menggunakan matriks korelasi (Pearson) untuk melihat seberapa kuat hubungan linear antar variabel numerik di dalam dataset. Nilai mendekati 1 menandakan korelasi positif yang kuat, mendekati -1 menandakan korelasi negatif yang kuat, dan 0 berarti tidak ada korelasi.
Beberapa temuan yang kami temukan:
1. Terdapat korelasi positif yang sangat tinggi (0.71) antara OTR dan cicilan. Hal ini sangat rasional dalam bisnis. Semakin mahal harga kendaraan (OTR), maka cicilannya berpotensi semakin tinggi juga.
2. Terdapat korelasi positif menengah (0.57) antara dp aktual dan OTR. Hal ini menunjukkan bahwa semakin mahal harga kendaraan cenderung menuntut setoran dp yang lebih besar.
3. Terdapat korelasi positif menengah di angka 0.37 antara DP aktual dan cicilan, yang mengindikasikan bahwa pelanggan yang membayar uang muka lebih besar umumnya tetap memiliki nominal angsuran yang tinggi karena mereka cenderung membeli kendaraan dengan harga (OTR) yang jauh lebih mahal.
4. Fitur umur nyaris tidak memiliki korelasi dengan variabel lainnya (berkisar antara -0.01 hingga -0.06). Hal ini menunjukkan bahwa profil usia tidak menentukan harga dan keputusan membeli motor.

Angka korelasi yang kini tampil sangat rasional dan sejalan dengan logika bisnis dunia nyata merupakan bukti matematis bahwa proses penanganan outlier sebelumnya telah berhasil.
